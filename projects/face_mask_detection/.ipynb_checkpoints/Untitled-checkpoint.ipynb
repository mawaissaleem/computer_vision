```python
from keras.layers import Conv2D, Conv2DTranspose, BatchNormalization, Add, Input
from keras.layers import Activation, Dense
from tensorflow.keras.layers import Conv2D, MaxPooling2D

def residual_upsampling_block(x, filters):
    # save the input for the skip connection
    residual = x
    
    # upsample using transposed convolution
    x = Conv2DTranspose(filters, kernel_size=(3, 3), strides=(2, 2), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    
    # apply convolutional layers
    x = Conv2D(filters, kernel_size=(3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(filters, kernel_size=(3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    
    # add the residual input
    x = Add()([x, residual])
    x = Activation('relu')(x)
    
    return x
```
